```python
import cv2
from canon.T4.utils import get_image_path
from pathlib import Path
from canon.config import BASE_MODEL_PATH
import numpy as np
from diffusers import StableDiffusionUpscalePipeline
from PIL import Image

import torch
print(torch.cuda.is_available())

pipe = StableDiffusionUpscalePipeline.from_pretrained(
    "stabilityai/stable-diffusion-x4-upscaler",
    torch_dtype=torch.float16
).to("cuda")

img_path = get_image_path("aleatoria")
lowres = Image.open(img_path).convert("RGB")

result = pipe(prompt="", image=lowres).images[0]
result.save("upscaled.png")
```
