```python
from bsrgan import BSRGAN
import cv2
from canon.T4.utils import get_image_path
from pathlib import Path
from canon.config import BASE_MODEL_PATH
import numpy as np

# Carrega imagem
img = get_image_path("antiga_2")

modelDir = Path(BASE_MODEL_PATH)
modelPath = modelDir / "BSRGAN.pth"

model = BSRGAN(
    model_path=str(modelPath),
    device="cpu",
    hf_model=False
)

# Agora passa o numpy corretamente
pred = model.predict(img_path=img)
print(f'Result here: {pred}')
```
